```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
life_expectancy_mean=df['Life Expectancy World Bank'].mean()

df['Life Expectancy World Bank'].fillna(life_expectancy_mean, inplace=True)
```

```python
undernourishment_median=df['Prevelance of Undernourishment'].median()

df['Prevelance of Undernourishment'].fillna(undernourishment_median, inplace=True)
```

```python
df=df.dropna(subset='CO2')
```

```python
Health_median=df['Health Expenditure %'].median()

df['Health Expenditure %'].fillna(Health_median, inplace=True,)
```

```python
Education_median= df['Education Expenditure %'].median()

df['Education Expenditure %'].fillna(Education_median, inplace=True,)
```

```python
df=df.dropna(subset='Unemployment')
```

```python
df = df.drop('Corruption', axis=1)
```

```python
sanitation_mean=df['Sanitation'].mean()

df['Sanitation'].fillna(sanitation_mean, inplace=True)
```
